# Library와 Framework 차이 (100자 내외)

Library는 사용자가 제어 흐름을 관리하며 필요한 기능을 선택하여 사용하는 반면, Framework는 제어 흐름을 스스로 관리하고 사용자는 그 안에서 필요한 기능을 작성하는 데 차이가 있다.

# 딥러닝과 머신러닝의 관계 및 특징, 차이 (200자 내외)

머신러닝은 컴퓨터가 학습할 수 있는 능력을 갖추게하는 기술의 집합으로, SVM, Decision Tree 등이 머신러닝 알고리즘에 해당한다.

딥러닝은 머신러닝의 하위 범주로, 인공 신경망에 기반한 기술이다. 컴퓨터 비전, 자연어 처리 등 여러 분야에서 활용된다.

일반적으로 딥러닝은 머신러닝보다 복잡한 작업을 수행할 수 있으며, 더 많은 데이터를 필요로 한다.
