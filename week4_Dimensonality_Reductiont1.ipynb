고유값이 큰 축 순서로 정렬하는 과정 + 주성분 개수 number을 고려해야한다는 점을 유의해서 함수를 정의해주세요

```python
from sklearn.preprocessing import StandardScaler

def MYPCA(X,number):
  scaler = StandardScaler()

  X_std=scaler.fit_transform(X)

  X_cov=np.cov(X_std,y=None, rowvar=True, bias=False, ddof=None, fweights=None, aweights=None,  dtype=None)

  eigenvalues, eigenvectors = np.linalg.eig(X_cov)
  eigenvalues.sort()

  # 고유값을 대각행렬로 변환
  lambda_matrix = np.diag(eigenvalues)

  # 고유벡터 행렬로 변환
  V = eigenvectors

  # 원래 공분산 행렬 복원
  reconstructed_X_cov = np.dot(np.dot(V, lambda_matrix), np.linalg.inv(V))

  return reconstructed_X_cov
```

```python
MYPCA(X,3)
```
