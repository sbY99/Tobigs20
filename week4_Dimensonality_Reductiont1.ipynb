ÏúÑÏùò Í≥ºÏ†ïÏùÑ ÌïòÎÇòÏùò Ìï®ÏàòÎ°ú ÌëúÌòÑÌï¥Ï£ºÏÑ∏Ïöî!

ÏúÑÏóêÏÑú ÎßåÎì† Ìï®Ïàò(new_coordinate)ÎèÑ Í∞ÄÏ†∏Ïò§ÏÖîÏÑú ÏÇ¨Ïö©Ìï¥ÎèÑ Ï¢ãÏïÑÏöî!

+ÌïòÏßÄÎßå Ïù¥ Ìï®ÏàòÎäî numberÏù¥ÎùºÎäî Ïù∏ÏûêÎ•º Î∞õÏïÑÏÑú pcaÏóêÏÑú Ï£ºÏÑ±Î∂Ñ Í∞úÏàòÎ•º Í≥†Î†§ÌïòÎèÑÎ°ù Ìï©ÎãàÎã§!

Í≥†Ïú†Í∞íÏù¥ ÌÅ∞ Ï∂ï ÏàúÏÑúÎ°ú Ï†ïÎ†¨ÌïòÎäî Í≥ºÏ†ï + Ï£ºÏÑ±Î∂Ñ Í∞úÏàò numberÏùÑ Í≥†Î†§Ìï¥ÏïºÌïúÎã§Îäî Ï†êÏùÑ Ïú†ÏùòÌï¥ÏÑú Ìï®ÏàòÎ•º Ï†ïÏùòÌï¥Ï£ºÏÑ∏Ïöî

```python
from sklearn.preprocessing import StandardScaler

def MYPCA(X,number):
  scaler = StandardScaler()

  X_std=scaler.fit_transform(X)

  X_cov=np.cov(X_std,y=None, rowvar=True, bias=False, ddof=None, fweights=None, aweights=None,  dtype=None)

  eigenvalues, eigenvectors = np.linalg.eig(X_cov)
  eigenvalues.sort()

  # Í≥†Ïú†Í∞íÏùÑ ÎåÄÍ∞ÅÌñâÎ†¨Î°ú Î≥ÄÌôò
  lambda_matrix = np.diag(eigenvalues)

  # Í≥†Ïú†Î≤°ÌÑ∞ ÌñâÎ†¨Î°ú Î≥ÄÌôò
  V = eigenvectors

  # ÏõêÎûò Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨ Î≥µÏõê
  reconstructed_X_cov = np.dot(np.dot(V, lambda_matrix), np.linalg.inv(V))

  return reconstructed_X_cov
```

```python
MYPCA(X,3)
```
