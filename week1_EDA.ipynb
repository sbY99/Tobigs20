# 이상치 확인

```python
# extract int type columns
int_cols = df.select_dtypes(include=['int64']).columns
int_cols_list = list(int_cols)
int_cols_list
```

```python
# chunking the columns which type is int64
from pandas._libs.hashtable import value_count
import copy

means = df[int_cols_list].mean()

group_dict = {}
check_list = []
for col in int_cols_list:
    if col in check_list:
        continue
    group_dict[col] = [other_col for other_col in df[int_cols_list].columns if abs(means[col] - means[other_col]) <= 30]
    for i in group_dict[col]:
        check_list.append(i)

group_list = []

for key, value in group_dict.items():
    if len(value) < 4:
        group_list.append(value)
    else:
        temp = copy.deepcopy(value)
        while len(temp) > 3:
            group_list.append(value[:3])
            temp = temp[3:]
        group_list.append(temp)


print(group_list)
print(len(group_list))
```

# 파생변수 생성

```python
def chunking_age(x):
    if x < 20 :
        return 0
    elif x < 30:
        return 1
    elif x < 40:
        return 2
    elif x < 60:
        return 3
    else :
        return 4
```

```python
df['age_chunk'] = df['Age'].apply(chunking_age)
```

```python
df['age_chunk'].value_counts().plot.pie()
plt.title("Age")
plt.show()
```

```python
statis_col_list = []
for i in int_cols_list:
    if 'Satis' in i:
        statis_col_list.append(i)

df['satisfaction'] = df[statis_col_list].apply(lambda x: int(x.mean()), axis = 1)
```
