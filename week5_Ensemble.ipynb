# Encoding

```python
# check variable
df['Avg_Utilization_Ratio'].value_counts()
```

```python
# for ordinal variables
Income_Category_map = {
    'Less than $40K' : 0,
    '$40K - $60K'    : 1,
    '$60K - $80K'    : 2,
    '$80K - $120K'   : 3,
    '$120K +'        : 4,
    'Unknown'        : 5
}


Card_Category_map = {
    'Blue'     : 0,
    'Silver'   : 1,
    'Gold'     : 2,
    'Platinum' : 3
}

Education_Level_map = {
    'Uneducated'    : 0,
    'High School'   : 1,
    'College'       : 2,
    'Graduate'      : 3,
    'Post-Graduate' : 4,
    'Doctorate'     : 5,
    'Unknown'       : 6
}

# y value
Attrition_Flag_map = {
    'Existing Customer' : 0,
    'Attrited Customer' : 1
}

df.loc[:, 'Income_Category'] = df['Income_Category'].map(Income_Category_map)
df.loc[:, 'Card_Category'] = df['Card_Category'].map(Card_Category_map)
df.loc[:, 'Attrition_Flag'] = df['Attrition_Flag'].map(Attrition_Flag_map)
df.loc[:, 'Education_Level'] = df['Education_Level'].map(Education_Level_map)
```

```python
from sklearn.preprocessing import LabelEncoder

lbe = LabelEncoder()

# encoding for norminal variables
cat_cols = [x for x in df.columns if df[x].dtype == 'object']

for c in cat_cols:
    df.loc[:, c] = lbe.fit_transform(df.loc[:, c])
```

```python
df.head()
```

```python
df = df.sample(frac = 1).reset_index(drop = True) #To shuffle data

X = df.drop(['Attrition_Flag', 'CLIENTNUM'], axis = 1)
y = df.Attrition_Flag
```

# Run Model

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score, log_loss, confusion_matrix
import numpy as np
```

```python
from sklearn.model_selection import StratifiedKFold

def run_model(X, y, model, k):

    kf = StratifiedKFold(n_splits = k)
    auc_scores = []
    i = 0

    for idx in kf.split(X = X, y = y):

        train_idx, val_idx = idx[0], idx[1]

        i += 1
        X_train = X.iloc[train_idx, :]
        y_train = y[train_idx]
        X_val = X.iloc[val_idx, :]
        y_val = y[val_idx]

        model.fit(X_train, y_train)
        y_predict = model.predict_proba(X_val)
        y_predict_prob = y_predict[:,1]

        auc_score = roc_auc_score(y_val, y_predict_prob)
        print(f'AUC Score of {i} Fold is : {auc_score}')
        auc_scores.append(auc_score)

    print('-----------------------------------------------')
    print(f'Average AUC Score of {k} Folds is : {np.mean(auc_scores)}')
```

```python
# Random Forest
params = {'n_estimators': 2000, 'max_depth': 20}
RandomForest = RandomForestClassifier(**params)

run_model(X, y, RandomForest, 3)
```

```python
# XGboost
params = {'n_estimators': 2000}
XGboost = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss', **params)

run_model(X, y, XGboost, 3)
```
